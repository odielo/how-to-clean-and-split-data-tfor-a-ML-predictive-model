```python
#in our 'size'column we noticed some M,gand K  symbols at the  end of each vaariable but 
#but we are going to use it in our model we came up with this function to to the convert 
#the sizes to one common unit ie MB 
def convert_size(size_str):
    if isinstance(size_str, float):
        return size_str
    elif size_str[-1] == 'K':
        return float(size_str[:-1]) / 1024
    elif size_str[-1] == 'M':
        return float(size_str[:-1])
    elif size_str[-1] == 'G':
        return float(size_str[:-1]) * 1024
    else:
        return size_str
data.head(10)
```

```python
#we now split our data into test,tain and validate 
from sklearn.model_selection import train_test_split

# split the data into train and test sets (70% train and 30% test)
train_data, test_data = train_test_split(data, test_size=0.3, random_state=42)

# split the test data into test and validation sets (20% test and 10% validation)
test_data, validation_data = train_test_split(test_data, test_size=0.33, random_state=42)

# print the shape of each set to confirm the split
print(f"Train set shape: {train_data.shape}")
print(f"Test set shape: {test_data.shape}")
print(f"Validation set shape: {validation_data.shape}")
# save the splits as CSV files
train_data.to_csv('play.train_data.csv', index=False)
test_data.to_csv('play.test_data.csv', index=False)
validation_data.to_csv('play.validation_data.csv', index=False)
```
